```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
import time
import random
import pandas as pd
import bs4
import requests

tickers = ['AMZN', 'AAPL', 'MSFT', 'GOOG', 'TSLA', 'JNJ', 'PG', 'NVDA', 'CSCO', 'BABA', 'HD',
'BIDU', 'WMT', 'CRM', 'LULU', 'TGT', 'PANW', 'ADBE', 'VMW', 'MU', 'NKE', 'ORCL', 'BB', 'HPQ', 'COST', 'AMAT',
'BAC', 'CVX', 'AMGN', 'PG']
tickers2 = ['TGT', 'PANW', 'ADBE', 'VMW', 'MU', 'NKE', 'ORCL', 'BB', 'HPQ', 'COST', 'AMAT',
'BAC', 'CVX', 'AMGN', 'PG']

url = 'https://estimize.com/'


# create a list to store data in
company_data = []
```

```python
# create path for driver location and setup driver services
path = r'C:\Users\brown\OneDrive\Documents\chromedriver.exe'
driver_service = Service(executable_path=path)
driver = webdriver.Chrome(service=driver_service)

for tick in tickers2:
    time.sleep(1)
    driver.get(f"{url}{tick}")
    time.sleep(random.uniform(2,6))

    # check if login needed
    try:
        login = driver.find_element('xpath','//*[@id="top-navigation"]/div/a[2]')
        login.click()
        time.sleep(1)

        # list username and password as variables
        username = 'cbrown180@student.gsu.edu'
        password = '123456'

        # find email field in HTML and send username to field -- sleep 2 seconds
        email_field = driver.find_element('name','user[login]')
        email_field.send_keys(username)
        time.sleep(1)

        # find the password field and input password and submit
        password_field = driver.find_element('name','user[password]')
        password_field.send_keys(password)
        password_field.submit()
        time.sleep(1)
    except NoSuchElementException:
        # create a dictonary to store data in
        overview = {}
        # find ticker 
        ticker = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/div[1]/h1/a').text
        overview['ticker'] = ticker
        # find company data
        company = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/div[1]/p/a').text
        overview['company'] = company
        # Sectors
        sector = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/p/span[1]/a/span').text
        overview['sector'] = sector
        # Industries
        industry = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/p/span[2]/a/span').text
        overview['industry'] = industry
        # number of folowers: 66
        followers = driver.find_element('xpath','//*[@id="summary-stats"]/div/div/div[1]/div[2]').text
        overview['followers'] = followers
        # number of analysts: 99
        analysts = driver.find_element('xpath','//*[@id="summary-stats"]/div/div/div[2]/a').text
        overview['analysts'] = analysts
        # add overview to list
        if overview not in company_data:
            company_data.append(overview)
    
    # create a dictonary to store data in
    overview = {}
    # find ticker 
    ticker = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/div[1]/h1/a').text
    overview['ticker'] = ticker
    # find company data
    company = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/div[1]/p/a').text
    overview['company'] = company
    # Sectors
    sector = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/p/span[1]/a/span').text
    overview['sector'] = sector
    # Industries
    industry = driver.find_element('xpath','//*[@id="releases_show"]/div[4]/div[2]/div/div[1]/div/div/div/p/span[2]/a/span').text
    overview['industry'] = industry
    # number of folowers: 66
    followers = driver.find_element('xpath','//*[@id="summary-stats"]/div/div/div[1]/div[2]').text
    overview['followers'] = followers
    # number of analysts: 99
    analysts = driver.find_element('xpath','//*[@id="summary-stats"]/div/div/div[2]/a').text
    overview['analysts'] = analysts
    # add overview to list
    if overview not in company_data:
        company_data.append(overview)
    time.sleep(random.uniform(1,3))



driver.quit()
```

```python

# store comapny data result in df and create csv file

df = pd.DataFrame(company_data)
print(df)
df.to_csv('companydata.csv')




  
```

```python
# testing for eps data
# create path for driver location and setup driver services
ticker_test = ['NVDA','AMZN']
path = r'C:\Users\brown\OneDrive\Documents\chromedriver.exe'
driver_service = Service(executable_path=path)
driver = webdriver.Chrome(service=driver_service)
eps_list = []

for ticks in ticker_test:
    
    time.sleep(1)
    driver.get(f"{url}{ticks}/fq2-2021?metric_name=eps&chart=table")
    time.sleep(random.uniform(1,1))
    # check if login needed
    try:
        login = driver.find_element('xpath','//*[@id="top-navigation"]/div/a[2]')
        login.click()
        time.sleep(1)
        # list username and password as variables
        username = 'cbrown180@student.gsu.edu'
        password = '123456'
        # find email field in HTML and send username to field -- sleep 2 seconds
        email_field = driver.find_element('name','user[login]')
        email_field.send_keys(username)
        time.sleep(1)
        # find the password field and input password and submit
        password_field = driver.find_element('name','user[password]')
        password_field.send_keys(password)
        password_field.submit()
        time.sleep(1)
    except NoSuchElementException:
        pass

    # find eps table data
    eps_header = driver.find_element('xpath','//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/thead').text
    quarters = eps_header.replace(" '",'_').split(' ')

    eps_estimize_c = driver.find_element('xpath','//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[2]/tr[1]').text
    est_consensus = eps_estimize_c.split(' ')

    eps_estimize_m = driver.find_element('xpath','//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[2]/tr[2]').text
    est_mean = eps_estimize_m.split(' ')

    eps_wallst = driver.find_element('xpath','//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[3]/tr[1]').text
    wallst_c = eps_wallst.split(' ')

    eps_actual = driver.find_element('xpath','//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[4]/tr[1]').text
    actual = eps_actual.split(' ')

    eps_dict = {'tick': f'{ticks}', 'quarters': quarters[1:],'estimize_consensus': est_consensus[1:], 'estimize_mean': est_mean[1:], 'wallst_consensus': wallst_c[1:], 'actual': actual[1:]}
    
    
    eps_list.append(eps_dict)

driver.quit()
```
