```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
import time
import random
import pandas as pd
import bs4
import requests
eps_list = []
```

#1 This code pulls pulls eps data for all tickers which have all quarters or all but q4'22 
and stores each quarterly results as a dictionary and appends eps_list as a row
 ======================================================================================= --!>


```python
# split group to prevent 404 error
# TODO: try script without ticker divide with google.get command. might be a fix
ticker_ok = ['BAC','CVX','AMGN']
ticker_ok2 = ['AMZN','GOOG','TSLA','JNJ']
ticker_ok3 = ['HD','NKE','BB','COST','PG','AAPL','MSFT','NVDA','BABA']

path = r'C:\Users\brown\OneDrive\Documents\chromedriver.exe'
driver_service = Service(executable_path=path)
driver = webdriver.Chrome(service=driver_service)
url = 'https://estimize.com/'
# eps_list = []

for ticks in ticker_ok3:
    driver.get('https://google.com')
    time.sleep(1)
    driver.get(f"{url}{ticks}/fq2-2021?metric_name=eps&chart=table")
    time.sleep(random.uniform(2,6))
    # check if login needed
    try:
        login = driver.find_element('xpath','//*[@id="top-navigation"]/div/a[2]')
        login.click()
        time.sleep(1)
        # list username and password as variables
        username = 'cbrown180@student.gsu.edu'
        password = '123456'
        # find email field in HTML and send username to field -- sleep 2 seconds
        email_field = driver.find_element('name','user[login]')
        email_field.send_keys(username)
        time.sleep(1)
        # find the password field and input password and submit
        password_field = driver.find_element('name','user[password]')
        password_field.send_keys(password)
        password_field.submit()
        time.sleep(1)
    except NoSuchElementException:
        pass

    
    for i in range(1,13):
        # create dictionary which will be row in total eps overview
        row = {}
        # get for each cat. and place in dict called row. append dict to total list to be a row in final product
        ticker = ticks
        row['ticker'] = ticker
        quarter = driver.find_element('xpath',f'//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/thead/tr/td[{i}]/a').text
        quarter = quarter.replace('F','').replace(' ','')
        time.sleep(random.uniform(1,2))
        row['quarter'] = quarter
        est_consensus = driver.find_element('xpath',f'//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[2]/tr[1]/td[{i}]').text
        row['e_consensus'] = est_consensus
        est_mean = driver.find_element('xpath',f'//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[2]/tr[2]/td[{i}]').text
        row['e_mean'] = est_mean
        wallst_consensus = driver.find_element('xpath',f'//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[3]/tr[1]/td[{i}]').text
        row['w_consensus'] = wallst_consensus
        actual_eps = driver.find_element('xpath',f'//*[@id="main-content"]/div[2]/div[3]/div[3]/div/table/tbody[4]/tr[1]/td[{i}]').text
                                              
        row['actual'] = actual_eps
        if row not in eps_list:
            eps_list.append(row)

      
   
```
